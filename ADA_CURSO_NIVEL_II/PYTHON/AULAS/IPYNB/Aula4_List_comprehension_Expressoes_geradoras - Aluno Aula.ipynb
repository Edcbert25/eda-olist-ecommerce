# List Comprehension

Suponha que voc√™ quer calcular o quadrado dos n√∫meros de uma lista, como far√≠amos com o que vimos at√© agora?

````python
numeros = [1, 2, 3, 4]
quadrados = []

for numero in numeros:
    quadrado = numero ** 2
    quadrados.append(quadrado)

quadrados
````

Uma list comprehension √© s√≥ uma forma mais curta e direta de escrever esse mesmo processo:

> pegar cada item da lista ‚Üí aplicar alguma transforma√ß√£o ‚Üí formar uma nova lista.

````python
[numero ** 2 for numero in numeros]
````

Estrutura b√°sica:


`[resultado for item in cole√ß√£o]`

- resultado ‚Üí o que quero gerar (ex: n**2)

- item ‚Üí cada elemento da lista original

- cole√ß√£o ‚Üí a lista ou sequ√™ncia que estou percorrendo

## List Comprehension

Compreens√£o de listas √© uma das maneiras mais pythonicas de se criar uma lista:<br>

```python
novaLista = [expressao for item in algoIteravel]
```

Tudo o que vamos ver em seguida pode ser aplicado a dicion√°rios tamb√©m.

Python, Haskell, Scala, Elixir, Erlang e Julia ‚Üí t√™m syntaxe nativa de list comprehension.

JavaScript e C# (LINQ) ‚Üí t√™m recursos equivalentes usando map e filter

Vamos comparar o tempo de execu√ß√£o de um c√≥digo com o `for` e o `list comprehension` e comparar usando a cell magic %%timeit:

````python
%%timeit
# criando uma lista de 0 a 100 a partir de for e append
lista_range = []
for n in range(0, 101):
    temp_f = n * 1.8 + 32
    lista_range.append(temp_f)

lista_range   
````

````python
%%timeit
# criando uma lista a partir de list comprehension de 0 a 100
lista_range = [n * 1.8 + 32 for n in range(0, 101)]

lista_range
````

Podemos peceber que o tempo de execu√ß√£o do list comprehension foi **menor** que o for tradicional.<br>
Ou seja, o list comprehension √© mais **otimizado** neste caso.

**OBS.:** Ao usar %%timeit em uma c√©lula do Jupyter Notebook, o c√≥digo contido na c√©lula √© executado v√°rias vezes e o tempo m√©dio de execu√ß√£o √© calculado. Isso ajuda a obter uma medi√ß√£o mais precisa e est√°vel do tempo de execu√ß√£o, j√° que fatores como flutua√ß√µes de carga do sistema podem afetar as medi√ß√µes individuais.

O IPython/Jupyter tem dois tipos de "magias":

- Line magics ‚Üí come√ßam com % e funcionam em uma linha. Exemplo: 
```sql
x = sum(range(1000))
%timeit x = sum(range(1000))

print(x)
```

- Cell magics ‚Üí come√ßam com %% e funcionam em toda a c√©lula.
