### **Usando o with:**

Uma outra forma de manipular arquivos sem se importar em sempre fech√°-lo √© utilizar `bloco de contexto`.<br>

Esse bloco controla automaticamente o ciclo de vida do recurso (nesse caso, o arquivo). Exemplo de uso:
<br><br>

#### Sem `with`

```python
arquivo = open('notas.txt', 'r', encoding='utf-8')
lista = arquivo.readlines()
arquivo.close()

```

#### Com `with`
```python
with open('arquivo.txt', 'r', encoding='utf-8') as arquivo:
    lista = arquivo.readlines()
```
<br>

Passo a passo do que acontece ao utilizarmos o with:

1. O Python abre o arquivo com `open()`

2. Entrega esse arquivo para a vari√°vel (`arquivo`)

3. Quando o bloco `with` termina (quando o recuo acaba), o Python chama automaticamente o m√©todo interno `arquivo.__exit__()`

4. Esse m√©todo executa arquivo.close() por voc√™ ‚Äî mesmo que ocorra um erro dentro do bloco. Caso contr√°rio poder√≠amos usar um try/except/finally.

Ou seja, utilizando o `with` o gerenciador chama e fecha conex√µes automaticamente. N√£o precisamos nos preocupar em explicitamente fechar conex√µes.

O `with` n√£o serve s√≥ pra arquivos ‚Äî ele funciona com qualquer objeto que implemente os m√©todos especiais
`__enter__()` e `__exit__()`.

Esses objetos s√£o chamados de `context managers` (gerenciadores de contexto).

Os mais comuns s√£o com arquivos, como o que vimos, e ao acessar bancos de dados:

```python
import sqlite3

with sqlite3.connect("meubanco.db") as conn:
    cursor = conn.cursor()
    cursor.execute("SELECT * FROM tabela")
```

### Escrita em arquivos csv com `csv.writer()`

Para inserir dados em um `csv` vamos precisar de quatro etapas:

1. Abrir o arquivo com `arquivo = open('nome_arquivo.csv', 'w', encoding='utf-8')`

2. `escritor = csv.writer()` para indicar onde as informa√ß√µes ser√£o escritas e de que forma 

```python

```

Indica:

- `arquivo`: objeto do arquivo no qual quer escrever
- `delimiter`: como que voc√™ vai separar os valores (neste caso, com ";")
- `lineterminator`: como vc vai separar as diferentes linhas do arquivo (neste caso, com "\n")

2. `escritor.writerows()` para indicar quais os dados que voc√™ quer escrever no arquivo (no caso, a lista de listas "tabela"):

```python
escritor.writerows(tabela)
```

3. `arquivo.close()` para fechar o arquivo. 

<br><br>

**Exerc√≠cio 1:**

Salve em um arquivo csv as notas de avalia√ß√£o dos filmes abaixo:

```python
Notas_filmes = [['Harry Potter', 10, 8, 9.5, 8.5, 10], ['Senhor dos An√©is', 9.7, 8.8, 10,
                8.9, 9.9], ['Velozes e Furiosos', 9, 8.1, 9.1, 7.5, 9.2], ['007', 8.7, 10, 8.5, 9.5, 9.1]]
```

````python
import csv
Notas_filmes = [['Harry Potter', 10, 8, 9.5, 8.5, 10], ['Senhor dos An√©is', 9.7, 8.8, 10,
                8.9, 9.9], ['Velozes e Furiosos', 9, 8.1, 9.1, 7.5, 9.2], ['007', 8.7, 10, 8.5, 9.5, 9.1]]

nome_arquivo = "notas_filmes.csv"

cabecalho = ['Filme', 'Nota_1', 'Nota_2', 'Nota_3', 'Nota_4', 'Nota_5']

with open(nome_arquivo, 'w', newline='', encoding='utf-8') as arquivo_csv:
    
    writer = csv.writer(arquivo_csv)
    
    writer.writerow(cabecalho)
    
    writer.writerows(Notas_filmes)
    
print(f" Dados salvos com sucesso no arquivo: **{nome_arquivo}**")
````

**Exerc√≠cio 2:**

Leia o arquivo salvo no exerc√≠cio 1 e retorne em um dicion√°rio a m√©dia de cada filme.


````python
media_filmes = {}

with open(nome_arquivo, 'r', newline='', encoding='utf-8') as arquivo_csv:
    reader = csv.reader(arquivo_csv, delimiter=',')

    next(reader)
    
    for linha in reader:
        nome_filme = linha[0]
        
        notas = [float(nota) for nota in linha[1:]]
        if notas:
            soma_notas = sum(notas)
            quantidade_notas = len(notas)
            
            if quantidade_notas > 0:
                media = soma_notas / quantidade_notas
                media_filmes[nome_filme] = round(media, 2)
                
for filme, media in media_filmes.items():
    print(f"{filme} | {media}")
````

**Exerc√≠cio 3:**

Salve em um arquivo txt em Python, com a frase: "O roeu a roupa do rei de Roma" e escreva um programa em python que recebe esse arquivo como entrada e retorne a quantidade de cada letra que aparece no texto sem fazer distin√ß√£o entre mai√∫sculas e min√∫sculas.

````python
def criar_arquivo():
    nome_arquivo = "texto.txt"
    
    frase = "o rato roeu a roupa do rei de Roma"
    
    arquivo = open(nome_arquivo, 'w', encoding='utf-8')
    
    arquivo.write(frase)
    
    arquivo.close()
    
    print(f"Arquivo '{nome_arquivo}' criado com sucesso contendo:\n{frase}'")
    
criar_arquivo()
````

````python
def contar_letras(nome_arquivo="texto.txt"):
    
    contagem_letras = {}
    
    arquivo = open(nome_arquivo, 'r', encoding='utf-8')
    
    conteudo = arquivo.read()
    
    arquivo.close()
    
    texto_processar = conteudo.lower()
    
    for caractere in texto_processar:
        if caractere.isalpha():
            contagem_letras[caractere] = contagem_letras.get(caractere, 0) + 1
    
    for letra in sorted(contagem_letras.keys()):
        quantidade = contagem_letras[letra]
        print(f"{letra}: {quantidade}")
        
contar_letras()
````
