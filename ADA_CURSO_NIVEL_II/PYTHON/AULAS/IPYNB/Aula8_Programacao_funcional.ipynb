# Funções Anônimas (Lambda):

Já imaginou precisar utilizar uma função, uma única vez, para um determinado propósito e não precisar usar mais? É neste caso que são utilizadas as funções anônimas, também conhecidas como **funções lâmbda ($\lambda$)**.

> Uma **função lambda** é uma **função anônima** — ou seja, sem nome — usada quando você **precisa de uma função pequena e rápida**, geralmente em uma única linha.

É como uma “mini função descartável”.

**Sintaxe Básica:**

```python
lambda parâmetros: retorno (expressao)
```

- `lambda` → palavra-chave que cria a função.

- `parâmetros` → o que a função recebe.

- `expressão` → o que ela retorna (sem precisar de return).

A própria estrutura da função lambda já nos informa quais parâmetros serão utilizados e já está implícito que retorna a expressão a frente.

Crie uma função lambda que receba a idade e retorne 'Maior' se >=18 e 'Menor' caso contrário


Você tem a lista: `idades = [15, 22, 17, 30, 19]`
Crie uma lista apenas com maiores de idade usando lambda


````python
idades = [15, 22, 17, 30, 19]
maior_idade = list(filter(lambda idade: idade > 18, idades))
print(maior_idade)
````

Você tem a lista: `precos = [100, 150, 200]`
Crie uma lista com preços acrescidos de 10% usando list comprehension e lambda

````python
precos = [100, 150, 200]
# lista_precos_lambda_map = list(map(lambda x: x * 1.1, precos))
# lista_precos_atualizada

lista_precos_lambda_comprehension = [(lambda x: x * 1.1)(x) for x in precos]
print(lista_precos_lambda_comprehension) 
````

Lista de nomes: `nomes = ['ana', 'carlos', 'beatriz']`
Crie uma nova lista com nomes em maiúsculas usando list comprehension e lambda

````python
nomes = ['ana', 'carlos', 'beatriz']
nomes_maiusculas = [(lambda nome: nome.upper())(n) for n in nomes]
print(nomes_maiusculas) 
````

Em um banco, um gerente de filial precisa calcular o saldo total de todas as contas dos clientes no final do mês para relatórios financeiros. <br> 
Ele deseja automatizar esse processo para economizar tempo e minimizar erros. Utilize a função reduce e compare com a função `sum`

```python
saldos = [1000, -500, 200, -300, 1500]
```

````python
from functools import reduce

saldos = [1000, -500, 200, -300, 1500]

saldo_total = reduce(lambda x, y: x + y, saldos)

print(saldo_total)
````
