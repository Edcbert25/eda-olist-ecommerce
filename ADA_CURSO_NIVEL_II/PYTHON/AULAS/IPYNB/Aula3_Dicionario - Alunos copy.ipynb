## Itera√ß√£o sobre dicion√°rios:
Suponha que tenha chegado o momento das bonifica√ß√µes na empresa e queremos saber quanto cada funcion√°rio vai receber no fim do m√™s. COnsidere uma bonifica√ß√£o de 10%.

###1.KEYS

````python
salario = {
    "27389267811": 5436.34,
    "09362893639": 4565.00,
    "46282902829": 6323.00
}
````

````python
bonificacao = {}
for key in salario.keys():
    bonificacao_cpf = salario[key]*1.1
    print(key, bonificacao_cpf)
    bonificacao[key] = bonificacao_cpf
bonificacao
````

### 2. values

````python
salario.values()

soma = 0
for value in salario.values():
    soma += value * 1.1
soma
````

### 3. items

````python
bonificacao = {}
for cpf, salario in salario.items():
    bonificacao[cpf] = salario*1.1
bonificacao
````

**Exerc√≠cio 5:**


Considere as seguintes idades:

```idades = {"Ana": 22, "Bruno": 27, "Carlos": 19, "Diana": 25}```

- Mostre a maior idade.

- Mostre o nome da pessoa mais velha.

- Calcule a m√©dia das idades.

````python
idades = {"Ana": 22, "Bruno": 27, "Carlos": 19, "Diana": 25}

maior_idade = max(idades.values())
print(f"A maior idade √©: {maior_idade}")

nome_mais_velho = max(idades, key=idades.get)
print(f"O nome da pessoa mais velha √©: {nome_mais_velho}")

soma_idades = sum(idades.values())

numero_pessoas = len(idades)

media_idades = soma_idades / numero_pessoas

print(f"A m√©dia das idades √©: {media_idades: .2f}")
````

**Exerc√≠cio 6:**

Escreva um programa que leia uma frase do usu√°rio e conte quantas vezes cada palavra aparece. Armazene os resultados em um dicion√°rio onde as keys s√£o as palavras e os values s√£o as contagens da palavra.

Exemplo:
`frase = "um tigre dois tigres tr√™s tigres"`

retorno:
`{'um': 1, 'tigre': 1, 'dois': 1, 'tigres': 2, 'tr√™s': 1}`

keys s√£o as palavras e os values s√£o as contagens da palavra.

````python
frase = input("Digite sua frase: ") 
def contar_frequencia_palavras(texto):
    
    lista_palavras = texto.split()

    contagem = {}
    for texto in lista_palavras:
        if texto in contagem:
            contagem[texto] += 1
        else:
            contagem[texto] = 1
    return contagem

resultado = contar_frequencia_palavras(frase)
print(resultado)
````

Dado um exemplo de dicion√°rio abaixo, crie uma fun√ß√£o que realiza o login de um usu√°rio, no qual as keys √© o login e os values √© a senha. O programa deve pedir o login de usuario e a senha, se estiver correto, retornar True e a mensagem de √™xito do login, se n√£o False e a mensagem de falha do login, indicando que n√£o foi poss√≠vel realizar o login. Al√©m disso, se o usu√°rio n√£o existir, tamb√©m retorna False e uma mensagem de que 'Senha incorreta ou usuario inexistente' (mas saberemos que isto significa que o usu√°rio n√£o existe rsrs).

```python
{
    "usuario1": "senha123",
    "usuario2": "abc456",
    "usuario3": "senhaforte",
    "john_doe": "d0e_p@ssw0rd",
    "alice": "al1ceP@ss"
}
```


````python
dados_usuarios = {
    "usuario1": "senha123",
    "usuario2": "abc456",
    "usuario3": "senhaforte",
    "john_doe": "d0e_p@ssw0rd",
    "alice": "al1ceP@ss"
}

login_digitado = input("Digite o login: ")
senha_digitada = input("Digite a senha: ")

def realizar_login(login, senha, base_dados):
    if login in base_dados:
        if base_dados[login] == senha:
            return True, f"Login de {login} realizado com sucesso"
        else:
            return False, f"Senha incorreta ou usu√°rio inexistente"
    else:
        return False, f"Senha incorreta ou usu√°rio inexistente"
    

status, mensagem = realizar_login(login_digitado, senha_digitada, dados_usuarios)


print(status)
print(mensagem)
````
