```python
import numpy as np
```

```python
a = (1, 2, 6)
b = (5, 2, 1)
c = (2, 2, 10)
```

```python
np.cross(a, b) # produto vetorial (axb)

np.abs(np.dot(c, np.cross(a, b))) # volume
```

```python
a = np.array([1, 2, 6])
b = np.array([5, 2, 1])
c = np.array([2, 2, 10])

np.abs(np.linalg.det(np.array([c, a, b])))
```
