```python
import numpy as np
```

```python
#base canônica

e1 = np.array([1, 0])
e2 = np.array([0, 1])
```

```python
#componentes dos vetores

v1, v2 = (4, 6)
w1, w2 = (3, 5)
```

```python
#vetores expandidos na base canônica

v = v1*e1 + v2*e2
w = w1*e1 + w2*e2
```

```python
# produto escalar, diretamente

(v*w).sum()
```

```python
#produto escalar, com a função no.dot()

np.dot(v, w)
```
