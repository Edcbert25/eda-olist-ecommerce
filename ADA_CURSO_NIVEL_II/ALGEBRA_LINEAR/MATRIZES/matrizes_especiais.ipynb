```python
### MATRIZ TRANSPOSTA - troca de linhas pelas colunas de uma matriz 
```

```python
import numpy as np
```

```python
matriz = np.array([[1, 1, -3], [2, 1, 3], [2, 0, -1]])
```

```python
matriz_t = matriz.transpose() # TRANSPONDO A MATRIZ
```

```python
matriz_t
```

```python
matriz.T    
```

```python
### MATRIZ INVERSA - obtida a partir de uma matriz quadrada, tal que o produto entre A e sua inversa (A^-1) resulta na matriz identidade (I).
# Determinante da matriz deve ser !=0
# Matriz deve ser quadrada (Anxn)
```

```python
A = matriz
A
```

```python
np.linalg.det(A) ### DETERMINANTE != 0 >>> TEM INVERSA!
```

```python
### CALCULANDO A MATRIZ INVERSA

A_inv = np.linalg.inv(A)
A_inv
```

```python
### CALCULANDO A MATRIZ IDENTIDADE

I = np.dot(A, A_inv)
I
```
