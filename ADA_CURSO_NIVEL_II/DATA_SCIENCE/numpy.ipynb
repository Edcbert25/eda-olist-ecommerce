```python
## Notas de alunos

lista_de_notas = [9.8, 5.6, 7.8, 9.1, 6.5]

notas = np.array(lista_de_notas)
```

```python
print("Maximo", notas.max())
print("Minimo", notas.min())
print("Desvio padrao", notas.std())
print("Media", notas.mean())
print("Posicao menor", notas.argmin())
```

```python
def calcula_notas():
    try:
        notas = (input("Digite as notas dos últimos 10 trabalhos: "))
        lista_notas = [float(n) for n in notas.split()]
        notas_arr = np.array(lista_notas)
    
        if len(notas_arr) == 0:
            return "Nenhuma nota inserida."
    
        return (f"\n---Estatísticas ---"
            f"\nMédia:          {notas_arr.mean():.2f}"
            f"\nDesvio Padrão:  {notas_arr.std():.2f}")
        
    except ValueError:
        return "Erro. Certifique-se de digitar apenas números separados por espaço."
```

```python
print(calcula_notas())
```
