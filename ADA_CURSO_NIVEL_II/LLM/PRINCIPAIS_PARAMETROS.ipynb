```python
# EXEMPLO COM MAX TOKENS

response = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[
        {"role":"system", "content": "Voc√™ √© um professor de uma escola infantil."},
        {"role":"user", "content": "Crie uma mensagem de boas-vindas aos alunos." }
    ],
    max_tokens=30
)
```

```python
print(response.choices[0].message.content)
```

```python
# TEMPERATURE: com valor alto(mais pr√≥ximo de 2), o modelo produz respostas mais criativas, diversas e, √†s vezes, surpreendentes. Ideal para tarefas que se beneficiam da criatividade.

response = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[
        {"role":"user", "content": "D√™ 5 sugest√µes de nome para um fundo de investimentos baseado em intelig√™ncia artifical." }
    ],
    temperature = 2
)
```

```python
print(response.choices[0].message.content)
```
