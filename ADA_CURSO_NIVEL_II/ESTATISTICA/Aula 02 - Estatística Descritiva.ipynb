## 1. Introdução

<br>

A __Estatística Descritiva__ é a área da estatística que busca conhecer e sintetizar as informações a partir de um conjunto de dados quaisquer. Existem diversas ferramentas estatísticas que podem ser utilizadas na interpretação dos dados, entre elas métricas, tabelas e gráficos que irão auxiliar no entendimento e ajudar a resumir as informações deste conjunto de dados. Um ponto importante é que para que essa análise estatística seja feita de forma clara e direta, deve-se entender os __tipos de variáveis__ e suas características para escolher as melhores abordagens.

---

## 2. Tipos de Variáveis

<br>

As variáveis são valores, numéricos ou não, que representam características de interesse a respeito do conjunto de dados. Na Estatística Descritiva, as variáveis mais utilizadas são separadas em duas categorias, sendo elas __qualitativas__ e __quantitativas__, onde, dentro destas categorias, pode-se dividir essas variáveis em dois grupos cada:

- __qualitativa nominal__: as variáveis do tipo _qualitativas_ não apresentam valores mensuráveis. No caso das variáveis __qualitativas__ e __nominais__, as variáveis __não apresentam uma ordenação ou hierarquia__ entre as categorias. __Exemplo:__ Sexo, País, estado civil e etc;


- __qualitativa ordinal__: Já para as variáveis __qualitativas__ e __ordinais__, as variáveis __apresentam uma ordenação ou hierarquia__ entre as categorias. __Exemplo:__ escolaridade, faixa salarial, período do dia e etc;


- __quantitativa discreta__: as variáveis do tipo _quantitativas_ apresentam valores mensuráveis, e para as variáveis __quantitativas__ e __discretas__, as variáveis são representadas por __quantidades enumeráveis__ (isto é, que podemos contar). Exemplos: Quantidade de filhos, quantidade de TVs, número de carros que trafegam por dia em  determinada rua, entre outros;


- __quantitativa contínua:__ as variáveis __quantitativas__ e __contínuas__ podem apresentar valores contínuos dentro da escala real, podendo apresentar valores fracionários, decimais, etc. Exemplo: Salário, fração de _Bitcoin_, altura e etc.

### 1.1 Medidas de Tendência Central

#### Conceito: Média
A **média aritmética** é o somatório de todos os valores dividido pela quantidade de observações.
- Interpretação prática: valor “esperado” de compra por cliente.
- Quando usar: dados sem valores extremos muito discrepantes.

#### Conceito: Mediana
A **mediana** é o valor que separa a distribuição ao meio (50 % acima, 50 % abaixo).
- Interpretação prática: ponto de equilíbrio, menos sensível a outliers.
- Quando usar: distribuições assimétricas.

### 1.2 Medidas de Dispersão e Forma

#### Conceito: Variância
A **variância** mede o quão dispersos estão os valores em relação à média:
$$
\sigma^2 = \frac{1}{n}\sum_{i=1}^{n}(x_i - \bar{x})^2
$$
- Quando usar: comparar consistência de diferentes produtos ou períodos.
- Aplicação no case ShopFast: uma variância alta demonstra que o ticket médio varia muito entre clientes, indicando necessidade de segmentar ofertas por perfil de gasto.

#### Conceito: Desvio-Padrão
O **desvio-padrão** é a raiz quadrada da variância:
$$
\sigma = \sqrt{\sigma^2}
$$
- Quando usar: comunicar dispersão ao time de marketing (ex.: “em média, as compras variam ± R$ 50”).
- Aplicação no case ShopFast: informa que a maior parte dos valores de compra fica dentro de um intervalo de ± desvio em torno da média, auxiliando na definição de faixas de preço para campanhas.

#### Conceito: Amplitude (Range)
A **amplitude** é a diferença entre o maior e o menor valor:
$$
\text{Range} = \max(x) - \min(x)
$$
- Quando usar: dimensionar categorias de preço.
- Aplicação no case ShopFast: mostra que os tickets vão de R$ X a R$ Y, ajudando a planejar limites mínimo e máximo de descontos.

#### Conceito: Coeficiente de Variação
O **coeficiente de variação** é o desvio-padrão dividido pela média:
$$
\text{CV} = \frac{\sigma}{\bar{x}}
$$
- Quando usar: avaliar consistência relativa de vendas entre diferentes produtos ou períodos.
- Aplicação no case ShopFast: permite comparar dispersões de tickets em categorias distintas (ex.: smartphones vs. acessórios), mesmo com médias diferentes.

#### Conceito: Assimetria e Curtose

A __assimetria__ é definida como o grau de desvio ou afastamento da simetria de uma distribuição. Quando determinada distribuição é simétrica, as métricas de posição como média, mediana e moda localizam-se em um mesmo ponto, havendo um perfeito equilíbrio na distribuição, mas normalmente este equilíbrio dentro de uma distribuição não acontece, a média, a mediana e a moda recaem em pontos diferentes da distribuição, e a distribuição é considerada __assimétrica__. Abaixo tem-se a representação das formas que uma distribuição qualquer pode apresentar com relação a simetria:

<img src="https://s3-sa-east-1.amazonaws.com/lcpi/618e3af5-bdcf-45f2-8ac4-47db139d857f.PNG" width=600>

<br>

Fonte: [Research Gate](https://www.researchgate.net/profile/Marco_Faria/publication/294890337/figure/fig5/AS:668873900716049@1536483373457/a-Negative-skewness-b-Normal-curve-c-Positive-skewness-Durkhure-and-Lodwal-2014.ppm)

<br>

Uma distribuição assimétrica pode ser classificada em 2 tipos de assimetrias:

- Distribuição assimétrica __Negativa__ ou __"à esquerda"__: as características das métricas para esta assimetria são que média < mediana < moda;


- Distribuição assimétrica __Positiva__ ou __à direita__: já para o caso da assimetria positiva tem-se que moda < mediana < média.

A métrica utilizada para avaliar o grau de assimetria de uma determinada distribuição é a __distorção__ (_skewness_), definido da seguinte forma:

<center><img align="center" src="https://i.upmath.me/svg/%20s(X)%20%3D%20%5Cfrac%7B1%7D%7B%5Csigma%5E3%7D%5Csum_%7Bi%3D1%7D%5E%7Bn%7D%5Cfrac%7B(x_i-%5Coverline%7Bx%7D)%5E%7B3%7D%7D%7Bn%7D" alt=" s(X) = \frac{1}{\sigma^3}\sum_{i=1}^{n}\frac{(x_i-\overline{x})^{3}}{n}" /></center>

Para identificar o tipo de assimetria a partir do _skewness_, deve-se seguir a referência de valores abaixo:

- s = 0: distribuição simétrica
- s > 0: assimetria à direita
- s < 0: assimetria à esquerda

<br>

A __Curtose__ é uma medida estatística que descreve o grau de achatamento ou "pico" de uma distribuição de dados em comparação com a curva normal, indicando a concentração de dados na média (pico) e nas caudas (valores extremos).


**Distribuições de probabilidade segundo a curtose**

- **Mesocúrtica**  
  - Curtose = 0 (igual à distribuição normal padrão).  
  - Apresenta caudas e pico de referência, sem achatamento ou concentração excessiva.  

- **Leptocúrtica**  
  - Curtose > 0.  
  - Pico mais alto e estreito, com caudas mais pesadas (MAIOR probabilidade de valores extremos).  

- **Platicúrtica**  
  - Curtose < 0.  
  - Pico mais baixo e achatado, com caudas mais leves (MENOR probabilidade de valores extremos).

## 2. Estatísticas Multivariadas

#### Covariância
A **covariância** mede como duas variáveis \(X\) e \(Y\) variam **conjuntamente**, ou seja, se quando uma aumenta, a outra tende a aumentar (covariância positiva) ou diminuir (covariância negativa).  

$$
\mathrm{Cov}(X,Y) 
= \frac{1}{n}\sum_{i=1}^{n}\bigl(x_i - \bar{x}\bigr)\,\bigl(y_i - \bar{y}\bigr)
$$

- **Unidade:** produto das unidades de \(X\) e \(Y\) (por exemplo, R\$·minutos).  
- **Interpretação:**  
  - $\mathrm{Cov}>0 $: em média, \(X\) e \(Y\) crescem juntos.  
  - $\mathrm{Cov}<0$: quando \(X\) cresce, \(Y\) tende a diminuir.  
  - $\mathrm{Cov}=0$: nenhuma tendência linear conjunta.

---

### Correlação

A **correlação** padroniza a covariância, removendo as unidades e restringindo o valor ao intervalo \([-1,1]\). O coeficiente de Pearson, por exemplo, é:

$$
r_{XY}
= \frac{\mathrm{Cov}(X,Y)}{\sigma_X\,\sigma_Y}
$$

- **Unidade:** adimensional (sem unidade).  
- **Intervalo:**  
  - $r\approx +1 $: forte associação linear positiva.  
  - $r\approx  0 $ pouca ou nenhuma associação linear.  
  - $r\approx -1 $: forte associação linear negativa.  

A **correlação de Pearson** padroniza a covariância, resultando em um valor entre –1 e 1.

- **Quando usar:** quantificar intensidade e direção de uma relação linear em variáveis quantitativas com distribuição aproximadamente normal.  
- **Aplicação no case ShopFast:** mensurar a relação linear entre `valor_compra` e `quantidade_itens` para avaliar cross-sell.

---



### Diferenças Principais

|                      | Covariância                          | Correlação                             |
|----------------------|--------------------------------------|----------------------------------------|
| **Escala**           | Depende das unidades de \(X\) e \(Y\)| Adimensional                          |
| **Intervalo**        | $(-\infty, +\infty)$               | \([-1, +1]\)                          |
| **Interpretação**    | Direção conjunta de variação         | Força **e** direção da relação linear |
| **Comparabilidade**  | Não comparável entre pares distintos | Comparável entre diferentes pares     |
| **Uso**              | Diagnosticar tendência bruta         | Quantificar intensidade e confiabilidade |

### Conceito: Correlação de Spearman  
A **correlação de Spearman** é uma medida não paramétrica entre as **posições (ranks)** das observações.  
- **Quando usar:** detectar relações monótonas (não necessariamente lineares) e em dados que não seguem normalidade.  
- **Aplicação no case ShopFast:** avaliar se clientes que passam mais tempo no site (`tempo_no_site`) tendem a gastar valores maiores, mesmo que a relação não seja perfeitamente linear.

```python
df.corr(method='spearman')

sns.heatmap(df.corr(method='spearman'), annot=True, cmap='viridis')
```

### Conceito: Correlação de Kendall’s Tau  
A **correlação de Kendall’s Tau** mede a **probabilidade de pares** de observações estarem em concordância ou discordância.  
- **Quando usar:** em amostras menores ou com muitos empates, fornece estimativa mais robusta que Spearman.  
- **Aplicação no case ShopFast:** entender a força da associação entre número de categorias diferentes compradas (`n_categorias`) e valor do ticket, quando há muitos empates no número de categorias.
