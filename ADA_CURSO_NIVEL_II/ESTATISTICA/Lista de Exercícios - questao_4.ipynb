```python
import numpy as np
np.random.seed(42)
populacao = np.random.normal(50, 15, 1000)
populacao = np.clip(populacao, 0, 100)

k= 10
inicio = np.random.randint(0, k)
amostra = populacao[inicio::k][0:100]

media_populacao = np.mean(populacao)
media_amostra = np.mean(amostra)

print(f'M√©dia de estresse (Popula√ß√£o): {media_populacao:.2f}')
print(f'M√©dia de estresse (Amostra): {media_amostra:.2f}')
print(f'Diferen√ßa (Erro Amostral): {abs(media_populacao - media_amostra):.2f}')
len(amostra)
```

```python
import pandas as pd
import numpy as np

Z = 1.96
p = 0.5
E = 0.05
N = 50_000

n_infinito = (Z**2) * p * (1 - p) / (E**2)

n_ajustado = n_infinito / (1 + (n_infinito) / N)

n_final = int(np.ceil(n_ajustado))

df_pop = pd.DataFrame({
    'investidor_id': np.arange(1, N + 1),
    'status_satisfeito': np.random.choice([0, 1], size=N, p=[0.5, 0.5])
})

amostra_df = df_pop.sample(n=n_final, random_state=42)

print(f'Tamanho da amostra calculado: {n_final}')
```
