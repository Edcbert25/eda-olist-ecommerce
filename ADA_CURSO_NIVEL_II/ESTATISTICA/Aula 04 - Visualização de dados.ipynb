```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# CONSTRUÇÃO DO DATASET
```

```python
rng = np.random.default_rng(42)
n = 800

df = pd.DataFrame({
    'idade': rng.integers(18, 70, n),
    'renda': rng.lognormal(mean=8.8, sigma=0.5, size=n),
    'score': rng.normal(loc=650, scale=80, size=n),
    'genero': rng.choice(['M', 'F'], size=n, p=[0.48, 0.52]),
    'segmento': rng.choice(['A', 'B','C'], size=n, p=[0.25, 0.5, 0.25])
    })
    
df
```

```python
### Definição de tema gráfico
```

```python
sns.set_theme(context='notebook', style='whitegrid')
```

```python
sns.histplot(df, x='renda', bins=10, kde=True)
plt.title('Histograma de renda')
plt.show()
```

```python

plt.figure()
sns.kdeplot(df, x='renda', fill=True)
plt.title('Histograma de renda')
plt.show()
```

```python
sns.kdeplot(df, x='idade', y='renda',fill=True, thresh=0.05, cmap='rocket_r')
```

```python
sns.pairplot(df, hue='segmento')
plt.show()
```

```python
plt.figure()
sns.jointplot(df, x='idade', y='renda', kind='scatter', color='teal')
plt.savefig('joint.png', dpi=100)
plt.show()
```
