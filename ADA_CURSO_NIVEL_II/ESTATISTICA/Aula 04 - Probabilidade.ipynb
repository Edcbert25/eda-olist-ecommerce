## 1. Conceitos Fundamentais
- **Probabilidade**: medida da chance de um evento ocorrer dado um conjunto de possibilidades, variando entre 0 e 1.  
- **Espa√ßo amostral (Œ©)**: conjunto de todos os resultados poss√≠veis de um experimento.  
- **Evento (E)**: subconjunto do espa√ßo amostral, representa o resultado de interesse. 

Exemplo:  
- Espa√ßo amostral ao lan√ßar um dado: Œ© = {1,2,3,4,5,6}  
- Evento de sair n√∫mero par: E = {2,4,6}  
- Probabilidade: P(E) = |E| / |Œ©| = 3/6 = 0.5  


A f√≥rmula cl√°ssica da probabilidade √©:

$$
P(A) = \frac{\text{n√∫mero de casos favor√°veis}}{\text{n√∫mero total de casos poss√≠veis}}
$$

Por exemplo, ao jogar um dado honesto de 6 faces, a chance de sair o n√∫mero 3 √©:

$$
P(3) = \frac{1}{6}
$$

## 2. Evento composto

Um **evento composto** √© formado pela combina√ß√£o de dois ou mais eventos simples dentro de um mesmo espa√ßo amostral.  
Ele pode ser constru√≠do por **uni√£o**, **interse√ß√£o** ou **complemento** de eventos.  

- **Uni√£o (A ‚à™ B)**: ocorre quando acontece A **ou** B.  
- **Interse√ß√£o (A ‚à© B)**: ocorre quando acontecem A **e** B simultaneamente.  
- **Complemento (A')**: ocorre quando o evento **n√£o acontece**.  

__Probabilidade de Eventos Compostos__

- **Uni√£o de eventos**:  
  P(A ‚à™ B) = P(A) + P(B) ‚Äì P(A ‚à© B)  

- **Interse√ß√£o de eventos independentes**:  
  P(A ‚à© B) = P(A) √ó P(B)  

- **Complemento**:  
  P(A') = 1 ‚Äì P(A)  

__Exemplo__

Se lan√ßarmos um dado:  
- Evento A = sair n√∫mero par = {2,4,6}  
- Evento B = sair n√∫mero maior que 3 = {4,5,6}  

- A ‚à™ B = {2,4,5,6}  
- A ‚à© B = {4,6}  
- A' = {1,3,5}

### üß™ Lab: BETS

- **Contexto:** Em jogos de azar online (bets) t√≠picos, a **probabilidade de ganhar** em uma rodada √© ligeiramente **menor** que a de perder, ou o **pagamento do ganho** √© um pouco **menor** do que a perda ‚Äî isso gera **valor esperado negativo** por rodada.
- **Ideia-chave:** Mesmo que o jogador acumule **vit√≥rias ocasionais**, com o passar das rodadas a **m√©dia do ganho acumulado** tende a **cair** (Lei dos Grandes N√∫meros), alinhando-se ao **drift negativo** igual ao **valor esperado por rodada √ó n√∫mero de rodadas**.
- **Modelo simples:** A cada rodada, apostar 1 unidade:
  - Probabilidade de vit√≥ria: `p_win = 0.48` (ganha `+1`)
  - Probabilidade de derrota: `1 - p_win` (perde `-1`)
  - Valor esperado por rodada: `E = p_win*1 - (1-p_win)*1 = 2*p_win - 1 < 0`

**Demonstra√ß√£o**
1. Simula milhares de jogadores por centenas/milhares de rodadas.
2. Calcula o **ganho acumulado** de cada jogador ao longo do tempo.
3. Plota:
   - **Linhas** de alguns jogadores (variabilidade individual).
   - **M√©dia** dos jogadores (converge para o drift negativo).
   - **Faixa P10‚ÄìP90** (percentis ao longo do tempo).
   - **Linha pontilhada** do **valor esperado** (`E √ó n_rodadas`).

---

```python
                             
# Simula√ß√£o de bets com valor esperado negativo e visualiza√ß√£o do ganho acumulado
import numpy as np
import matplotlib.pyplot as plt

# -----------------------
# Par√¢metros do experimento
# -----------------------
np.random.seed(42)   # reprodutibilidade
n_players = 2000     # n√∫mero de jogadores independentes
n_rounds = 1200      # n√∫mero de rodadas (tempo)
p_win = 0.48         # probabilidade de ganhar a rodada
win_payout = 1.0     # ganho por vit√≥ria (em unidades de stake)
loss_amount = 1.0    # perda por derrota
stake = 1.0          # aposta por rodada

# Valor esperado por rodada (drift): negativo em jogos desvantajosos
expected_gain_per_round = p_win * win_payout - (1 - p_win) * loss_amount

# -----------------------
# Simula√ß√£o
# -----------------------
# Matriz de resultados: 1 = vit√≥ria, 0 = derrota (shape: n_players x n_rounds)
wins = (np.random.rand(n_players, n_rounds) < p_win).astype(float)

# Lucro por rodada: +stake quando vence, -stake quando perde
profit_round = wins * win_payout * stake - (1 - wins) * loss_amount * stake

# Ganho acumulado ao longo das rodadas, por jogador
cum_profit = np.cumsum(profit_round, axis=1)  # (n_players, n_rounds)

# Estat√≠sticas agregadas ao longo dos jogadores
mean_cum_profit = cum_profit.mean(axis=0)
p10 = np.percentile(cum_profit, 10, axis=0)
p90 = np.percentile(cum_profit, 90, axis=0)

# Seleciona alguns jogadores para ilustrar trajet√≥rias individuais
example_indices = np.linspace(0, n_players-1, 6, dtype=int)
examples = cum_profit[example_indices, :]

# -----------------------
# Gr√°fico
# -----------------------
plt.figure(figsize=(10, 6))
x = np.arange(1, n_rounds + 1)

# Trajet√≥rias de alguns jogadores (linhas finas)
for row in examples:
    plt.plot(x, row, linewidth=0.8, alpha=0.6)

# M√©dia dos jogadores (linha mais espessa)
plt.plot(x, mean_cum_profit, linewidth=2.0, label="M√©dia dos jogadores")

# Tend√™ncia te√≥rica (valor esperado acumulado)
plt.plot(x, x * expected_gain_per_round, linestyle='--', linewidth=1.5, label="Tend√™ncia (valor esperado)")

# Faixa P10‚ÄìP90
import numpy as np  # redundante, mas mantido para clareza de execu√ß√£o isolada
plt.fill_between(x, p10, p90, alpha=0.15, label="Faixa P10‚ÄìP90")

plt.title("Jogo com Valor Esperado Negativo: ganho acumulado ao longo do tempo")
plt.xlabel("Rodadas (tempo)")
plt.ylabel("Ganho acumulado (unidades de aposta)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
